```python
# principial component analisis
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("seaborn-whitegrid")
```

```python
iris = load_iris()
iris
```

```python
data = iris.data
feature_names = iris.feature_names
y = iris.target

df = pd.DataFrame(data, columns = feature_names)
df["sinif"] = y

df.head()
```

```python
from sklearn.decomposition import PCA
x = data
pca = PCA(n_components=2, whiten= True)  # whitten = normalize
pca.fit(x)

x_pca = pca.transform(x)
print("veriance ratio: ",pca.explained_variance_ratio_)
print("sum: ",sum(pca.explained_variance_ratio_))
```

```python
# 2D görselleştirme
df["p1"] = x_pca[:,0]
df["p2"] = x_pca[:,1]


for each in range(3):
    plt.scatter(df.p1[df.sinif == each],df.p2[df.sinif == each],label = iris.target_names[each])
plt.legend()
plt.show()
```
